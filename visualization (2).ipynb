5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
# Generate random data
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 * 0.8 + np.random.rand(n_samples) * 0.2  # Correlated with x1
x3 = x1 * -0.5 + np.random.rand(n_samples) * 0.5  # Inversely correlated with x1
x4 = np.random.rand(n_samples)  # Uncorrelated feature

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Step 2: Calculate the correlation matrix
correlation_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Step 4: Add titles
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean', fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5)

# Step 4: Add titles and labels
plt.title('Heatmap of Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function x = np.linspace(-5, 5, 100)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Step 2: Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Step 3: Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Step 4: Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(df['X'].values.reshape(x.shape), df['Y'].values.reshape(y.shape), df['Z'].values.reshape(z.shape),
                cmap='viridis', edgecolor='none')

# Step 5: Add titles and labels
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Step 6: Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5, c='blue', edgecolors='w', linewidth=2)

plt.title('Bubble Chart of Countries: Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=10, ha='right')

plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Display plot in notebook
output_notebook()

# Create x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create Bokeh figure
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Plot the sine wave
p.line(x, y, line_width=2, color='blue')

# Show the plot
show(p)
```
